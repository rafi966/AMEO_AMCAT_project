```python
#converting the 10board as icse ,cbse and state board
board_10 = []
for i in df['10board']:
    if i == 'cbse' or i == 'all india board' or i=='central board of secondary education, new delhi' or i=='cbese':
        board_10.append('cbse')

    elif i=='isc' or i=='icse' or i=='isc board' or i=='isce' or i=='cicse' or i=='isc board , new delhi':
        board_10.append('icse')

    else:
        board_10.append('state board')

df['10board'] = board_10
```

```python
#converting the 12board as icse ,cbse and state board


board_12 =[]
for i in df['12board']:
    if i == 'cbse' or i == 'all india board' or i=='central board of secondary education, new delhi' or i=='cbese':
        board_12.append('cbse')

    elif i=='isc' or i=='icse' or i=='isc board' or i=='isce' or i=='cicse' or i=='isc board , new delhi':
        board_12.append('icse')

    else:
        board_12.append('state board')

df['12board'] = board_12
```
